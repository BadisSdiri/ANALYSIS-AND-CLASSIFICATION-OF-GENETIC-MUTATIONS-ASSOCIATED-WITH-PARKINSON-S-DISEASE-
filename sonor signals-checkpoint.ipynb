```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from scipy.io.wavfile import write
import os
```

```python
def nucleotide_to_freq(nucleotide):
    mapping = {'A': 440.0,  
               'T': 494.0,  
               'C': 523.0,  
               'G': 587.0}  
    return mapping.get(nucleotide, 440.0) 
```

```python
def generate_sound(sequence, duration=0.1, sample_rate=44100):
    signal = np.array([], dtype=np.float32)
    for nucleotide in sequence:
        freq = nucleotide_to_freq(nucleotide)
        t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
        wave = 0.5 * np.sin(2 * np.pi * freq * t)
        signal = np.concatenate([signal, wave])
    return signal
```

```python
def save_sound_files(data, output_dir):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    for index, row in data.iterrows():
        sequence = row['sequence']
        label = row['label']
        signal = generate_sound(sequence)
        file_name = f"{label}_{index}.wav"
        file_path = os.path.join(output_dir, file_name)
        write(file_path, 44100, signal.astype(np.float32))
```

```python
import os
import pandas as pd
import numpy as np
import librosa
def extract_features(file_path):
    y, sr = librosa.load(file_path, sr=None)  
    features = {}
    features['mfcc'] = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13).mean(axis=1)  
    features['spectral_centroid'] = librosa.feature.spectral_centroid(y=y, sr=sr).mean()  
    features['spectral_bandwidth'] = librosa.feature.spectral_bandwidth(y=y, sr=sr).mean()  
    features['zero_crossing_rate'] = librosa.feature.zero_crossing_rate(y).mean()  
    features['chroma_stft'] = librosa.feature.chroma_stft(y=y, sr=sr).mean(axis=1)  
    
    return features
def process_directory(directory):
    data = []
    for subdir, _, files in os.walk(directory):
        label = os.path.basename(subdir)  
        for file in files:
            if file.endswith(".wav"):
                file_path = os.path.join(subdir, file)
                features = extract_features(file_path)
                row = {f'mfcc_{i}': mfcc for i, mfcc in enumerate(features['mfcc'])}
                row.update({f'chroma_{i}': chroma for i, chroma in enumerate(features['chroma_stft'])})
                row['spectral_centroid'] = features['spectral_centroid']
                row['spectral_bandwidth'] = features['spectral_bandwidth']
                row['zero_crossing_rate'] = features['zero_crossing_rate']
                row['label'] = label
                data.append(row)
    return pd.DataFrame(data)

training_sound_dir = r"C:\Users\Badis\Desktop\training_sounds_separatedx"
test_sound_dir = r"C:\Users\Badis\Desktop\test_sounds_separatedx"

training_features_df = process_directory(training_sound_dir)
test_features_df = process_directory(test_sound_dir)

training_features_df.to_csv('training_features.csv', index=False)
test_features_df.to_csv('test_features.csv', index=False)

print("Feature extraction complete and saved to CSV.")
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

def plot_spectrogram(file_path):
    y, sr = librosa.load(file_path, sr=None)
    S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
    S_db = librosa.power_to_db(S, ref=np.max)
    plt.figure(figsize=(10, 6))
    librosa.display.specshow(S_db, sr=sr, x_axis='time', y_axis='mel', cmap='magma')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Mel-Spectrogram')
    plt.tight_layout()
    plt.show()
file_path = r"C:\Users\Badis\Desktop\training_sounds_separated\normal\normal_0.wav"
plot_spectrogram(file_path)
```
